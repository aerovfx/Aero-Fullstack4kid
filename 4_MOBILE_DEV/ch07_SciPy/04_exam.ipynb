**Giải phương trình vi phân**: SciPy có các hàm để giải các phương trình vi phân thông thường và phương trình vi phân đạo hàm riêng.

```python

import numpy as np
from scipy import integrate
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Định nghĩa hàm mô tả phương trình vi phân
def model(y, t):
    dydt = -y
    return dydt

# Điều kiện ban đầu
y0 = 5

# Tạo mảng thời gian
t = np.linspace(0, 10, 100)

# Giải phương trình vi phân
y = odeint(model, y0, t)

# Vẽ đồ thị kết quả
plt.plot(t, y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.title('Giải phương trình vi phân: dy/dt = -y')
plt.show()
```
