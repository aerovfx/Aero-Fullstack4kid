 **Biểu đồ phân loại**: Thư viện này hỗ trợ các biểu đồ phân loại như `boxplot`, `violinplot`, và `stripplot`.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Dữ liệu mẫu
data = sns.load_dataset('tips')

# Biểu đồ Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=data)
plt.title('Biểu đồ Boxplot của Tổng Hóa Đơn theo Ngày')
plt.show()

# Biểu đồ Violinplot
plt.figure(figsize=(10, 6))
sns.violinplot(x='day', y='total_bill', data=data)
plt.title('Biểu đồ Violinplot của Tổng Hóa Đơn theo Ngày')
plt.show()

# Biểu đồ Stripplot
plt.figure(figsize=(10, 6))
sns.stripplot(x='day', y='total_bill', data=data)
plt.title('Biểu đồ Stripplot của Tổng Hóa Đơn theo Ngày')
plt.show()
```
