2. **Biểu đồ quan hệ**: Seaborn có thể vẽ các biểu đồ quan hệ giữa các biến như `scatterplot`, `lineplot`, và `relplot`.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Tạo dữ liệu mẫu
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [5, 4, 3, 2, 1]
})

# Vẽ scatterplot
sns.scatterplot(data=data, x='x', y='y')
plt.title('Scatterplot giữa x và y')
plt.show()

# Vẽ lineplot
sns.lineplot(data=data, x='x', y='y')
plt.title('Lineplot giữa x và y')
plt.show()

# Vẽ relplot
sns.relplot(data=data, x='x', y='y', kind='scatter')
plt.title('Relplot giữa x và y')
plt.show()
```
