**Biểu đồ hồi quy**: Seaborn cung cấp các hàm như `regplot` và `lmplot` để vẽ các biểu đồ hồi quy.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Tạo dữ liệu mẫu
np.random.seed(0)
x = np.random.rand(100)
y = 2 * x + 1 + np.random.randn(100) * 0.1
data = pd.DataFrame({'x': x, 'y': y})

# Vẽ biểu đồ hồi quy sử dụng regplot
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.regplot(x='x', y='y', data=data)
plt.title('Biểu đồ hồi quy sử dụng regplot')

# Vẽ biểu đồ hồi quy sử dụng lmplot
plt.subplot(1, 2, 2)
sns.lmplot(x='x', y='y', data=data)
plt.title('Biểu đồ hồi quy sử dụng lmplot')

plt.tight_layout()
plt.show()
```
