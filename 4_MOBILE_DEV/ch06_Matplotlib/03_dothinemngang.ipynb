```python
import matplotlib.pyplot as plt
import numpy as np

# Các hằng số
g = 9.81  # gia tốc trọng trường (m/s^2)
v0 = 20  # vận tốc ban đầu (m/s)
theta = 0  # góc phóng (độ)
y0 = 10  # chiều cao ban đầu (m)

# Thời gian bay
t_flight = np.sqrt(2 * y0 / g)

# Khoảng thời gian
t = np.linspace(0, t_flight, num=500)

# Khoảng cách ngang
x = v0 * t

# Khoảng cách dọc
y = y0 - 0.5 * g * t**2

# Vẽ đồ thị quỹ đạo
plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.title('Đồ thị ném ngang')
plt.xlabel('Khoảng cách ngang (m)')
plt.ylabel('Khoảng cách dọc (m)')
plt.grid(True)
plt.show()
```
