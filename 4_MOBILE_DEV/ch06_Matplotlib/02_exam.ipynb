```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu mẫu
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Tạo một figure và axis
fig, ax = plt.subplots()

# Vẽ dữ liệu với tùy chỉnh
ax.plot(x, y, color='purple', linestyle='--', linewidth=2, marker='o', markerfacecolor='red', markersize=8)

# Thêm nhãn và tiêu đề
ax.set_xlabel('Nhãn trục X')
ax.set_ylabel('Nhãn trục Y')
ax.set_title('Ví dụ biểu đồ tùy chỉnh')

# Thêm lưới
ax.grid(True)

# Hiển thị biểu đồ
plt.show()
```
