```python
import matplotlib.pyplot as plt
"""
Vẽ Đồ Thị Đa Dạng với Matplotlib

Chương trình này minh họa cách vẽ các loại đồ thị khác nhau sử dụng thư viện Matplotlib bao gồm:
- Đồ thị đường (Line plot)
- Đồ thị cột (Bar plot)
- Đồ thị phân tán (Scatter plot)
- Biểu đồ histogram (Histogram)
- Đồ thị 3D (3D plot)

Các phần chính của chương trình:
1. Khởi tạo dữ liệu cho các đồ thị.
2. Vẽ đồ thị đường với hàm sin(x) và cos(x).
3. Vẽ đồ thị cột với các danh mục và giá trị tương ứng.
4. Vẽ đồ thị phân tán với các điểm ngẫu nhiên.
5. Vẽ biểu đồ histogram với dữ liệu ngẫu nhiên.
6. Vẽ đồ thị 3D với hàm sin của khoảng cách từ gốc tọa độ.

Các thư viện sử dụng:
- matplotlib.pyplot: Thư viện chính để vẽ đồ thị.
- numpy: Thư viện để xử lý mảng và tạo dữ liệu ngẫu nhiên.
- mpl_toolkits.mplot3d: Thư viện để vẽ đồ thị 3D.
"""
import numpy as np

# Data for plotting
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
data = np.random.randn(1000)

# Line plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Line Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# Bar plot
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 40]

plt.figure(figsize=(10, 6))
plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# Scatter plot
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)

plt.figure(figsize=(10, 6))
plt.scatter(x_scatter, y_scatter)
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, alpha=0.7, color='g')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# 3D plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
x_3d = np.linspace(-5, 5, 100)
y_3d = np.linspace(-5, 5, 100)
x_3d, y_3d = np.meshgrid(x_3d, y_3d)
z_3d = np.sin(np.sqrt(x_3d**2 + y_3d**2))

ax.plot_surface(x_3d, y_3d, z_3d, cmap='viridis')
ax.set_title('3D Plot')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```
