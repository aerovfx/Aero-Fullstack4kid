```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Tạo một tập dữ liệu đơn giản bằng NumPy
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Tạo một DataFrame bằng Pandas
df = pd.DataFrame({'x': x, 'y': y})

# Vẽ dữ liệu bằng Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(df['x'], df['y'], label='sin(x)')

# Thêm các đường nét đứt cho trục x
for xc in np.arange(0, 11, 1):
    plt.axvline(x=xc, color='r', linestyle='--', linewidth=0.5)

plt.title('Biểu đồ đơn giản sử dụng Matplotlib, NumPy và Pandas')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```
