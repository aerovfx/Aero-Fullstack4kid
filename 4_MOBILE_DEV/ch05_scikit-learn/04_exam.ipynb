```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Tải bộ dữ liệu Iris
iris = load_iris()
X = iris.data
y = iris.target

# Áp dụng PCA để giảm số chiều xuống còn 2 thành phần
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Vẽ đồ thị dữ liệu đã giảm số chiều
plt.figure(figsize=(8, 6))
for target in np.unique(y):
    plt.scatter(X_reduced[y == target, 0], X_reduced[y == target, 1], label=iris.target_names[target])
plt.xlabel('Thành phần chính 1')
plt.ylabel('Thành phần chính 2')
plt.legend()
plt.title('PCA của bộ dữ liệu Iris')
plt.show()
```
